```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split # 학습, 테스트set 구분
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd
```

```python
df = pd.read_csv('preprocessing1.csv', encoding = 'cp949', index_col=0 )
df
```

```python
seed = 5764

X=df.drop('Loan Status',axis=1)
Y=df['Loan Status']

# 학습, 테스트 데이터 분리 (0.7:0.3)
x_train, x_valid, y_train, y_valid = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
rf= RandomForestClassifier(n_estimators=10)

#fit model
rf= rf.fit(x_train,y_train)
#predict
pred= rf.predict(x_valid)
```

```python
confusion_matrix(y_valid,pred)
```

```python
accuracy_score(y_valid,pred)
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import recall_score
print("Accuracy of model ",accuracy_score(y_valid, pred))
print("F1 Score ",f1_score(y_valid, pred))
print("Recall Score ",recall_score(y_valid, pred))
print("Balanced Accuracy Score ",balanced_accuracy_score(y_valid, pred))
```
